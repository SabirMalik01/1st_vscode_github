**A clear and categorized list** of the most commonly used **libraries in Data Science.**

---

# 🔢 **1. Data Manipulation & Analysis**

### **Pandas**

* Used for working with **tables**, data cleaning, filtering, merging.
* Works with DataFrames (rows & columns).

### **NumPy**

* Performs **fast mathematical operations**.
* Used for arrays, matrices, linear algebra.

### **Dask**

* Like pandas but for **big datasets** that don’t fit in memory.

---

# 📊 **2. Data Visualization**

### **Matplotlib**

* Base plotting library.
* Creates line charts, bar charts, histograms, etc.

### **Seaborn**

* Built on top of Matplotlib.
* Creates **attractive, statistical** graphics easily.

### **Plotly**

* Creates **interactive** graphs (zoom, hover, rotate).

### **Bokeh**

* For interactive dashboards and visualizations.

---

# 🤖 **3. Machine Learning**

### **Scikit-learn**

* Most popular ML library.
* For classification, regression, clustering, preprocessing, model evaluation.

### **TensorFlow**

* Deep learning library from Google.
* Used for neural networks, CNNs, NLP.

### **Keras**

* High-level deep learning API running on TensorFlow.
* Very easy to build neural networks.

### **PyTorch**

* Deep learning library from Meta (Facebook).
* Very popular in research and AI models.

### **XGBoost**

* For high-performance boosting models.
* Excellent for competitions and tabular data.

### **LightGBM / CatBoost**

* Modern, fast gradient boosting libraries.

---

# 🧪 **4. Data Cleaning & Preprocessing**

### **OpenRefine (not Python, but used in DS)**

* Cleaning large messy datasets.

### **Regex (re module)**

* Text cleaning and pattern matching.

---

# 📦 **5. Big Data & Distributed Computing**

### **PySpark**

* Python API for **Apache Spark**.
* Used for massive datasets and distributed systems.

### **Hadoop (via APIs)**

* For big data storage and processing.

---

# 🛠️ **6. Natural Language Processing (NLP)**

### **NLTK**

* Basic NLP tasks (tokenization, stemming).

### **spaCy**

* Faster, industrial NLP processing.

### **Transformers (Hugging Face)**

* Modern NLP (BERT, GPT, T5).

---

# 📐 **7. Mathematics & Statistics**

### **SciPy**

* Advanced scientific computing (optimization, statistics).

### **Statsmodels**

* Detailed statistical modeling (ANOVA, linear models).

---

# 🌐 **8. Web Scraping & Data Collecting**

### **BeautifulSoup**

* Scrapes HTML content.

### **Scrapy**

* Powerful web-scraping framework.

### **Requests**

* For downloading web pages and APIs.

---

# 🗃️ **9. Databases**

### **SQLAlchemy**

* Python ORM for databases.

### **PyMySQL, psycopg2**

* Connect Python to MySQL/PostgreSQL.

---

# 🧹 **10. Data Engineering / Pipelines**

### **Airflow**

* Workflow automation (ETL pipelines).

### **Prefect / Luigi**

* Manage data pipelines and tasks.

---
