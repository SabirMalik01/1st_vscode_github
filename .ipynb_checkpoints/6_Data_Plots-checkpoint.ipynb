---

## Built-in Seaborn styles

### 1. **darkgrid** (Most popular)

* Dark grid on light background
* Best for **data analysis**

```python
sns.set_style("darkgrid")
```

---

### 2. **whitegrid**

* Light grid on white background
* Good for **categorical data**

```python
sns.set_style("whitegrid")
```

---

### 3. **dark**

* Dark background, no grid
* Good for **presentations**

```python
sns.set_style("dark")
```

---

### 4. **white**

* Clean white background, no grid
* Best for **minimal visuals**

```python
sns.set_style("white")
```

---

### 5. **ticks**

* White background with ticks
* Looks similar to scientific plots

```python
sns.set_style("ticks")
```

---

## Compare styles quickly

```python
sns.axes_style()
```

Different palettes in Seaborn :
- Set1, Set2, Set3
  
- pastel

- deep

- muted

- coolwarm

- viridis

  Palettes

  A palette is a set of colors used to represent categories or values in a plot 🎨

How to use a palette
- sns.barplot(x="species", y="sepal_length", data=flower, palette="Set2")

Types of palettes in Seaborn
1️⃣ Qualitative palettes (for categories)

Used when data is categorical

Common ones:

deep (default)

muted

pastel

bright

dark

colorblind

Set1, Set2, Set3

sns.set_palette("Set1")

2️⃣ Sequential palettes (for ordered data)

Used when values go from low → high

Examples:

Blues

Greens

Reds

Purples

Oranges

viridis

magma

plasma

sns.heatmap(data, cmap="Blues")

3️⃣ Diverging palettes (for positive & negative values)

Used when data has a center point (0, mean)

Examples:

coolwarm

RdBu

Spectral

vlag

icefire

sns.heatmap(data, cmap="coolwarm")

How to see palettes
sns.color_palette("Set2")


To display:

sns.palplot(sns.color_palette("Set2"))

Create custom palette
my_colors = ["red", "blue", "green"]
sns.set_palette(my_colors)

Set palette globally
sns.set_palette("pastel")

*Palette vs color* :
Feature	color	palette
- Single color	✅	❌
- Multiple colors	❌	✅
- Best for categories	❌	✅

### >>>Applying Parameters
Parameters control the appearence of bars and error bars in plots.
- linewidth(The width of bar lines)
- errcolor(The color of arrow bars)(0 for black and 1 for white)
- edgecolor(The color of bar lines)(0 for black and 1 for white)
- facecolor(x1,x2,x3,x4)

## 1️⃣ `linewidth`

```python
linewidth=3
```

### What it does

* Controls the **thickness of the bar borders (edges)**

### Meaning

* Higher value → thicker border
* Lower value → thinner border

📌 `linewidth=3` means **thick bar outlines**

---

## 2️⃣ `facecolor`

```python
facecolor=(1, 1, 1, 1)
```

### What it does

* Sets the **fill color inside the bars**

### The tuple `(1,1,1,1)` means (RGBA):

| Value | Meaning         |
| ----- | --------------- |
| 1     | Red             |
| 1     | Green           |
| 1     | Blue            |
| 1     | Alpha (opacity) |

➡️ `(1,1,1,1)` = **white color, fully opaque**

📌 So your bars are **white inside**
- 0 for no intensity
- 1 for full intensity

---

## 3️⃣ `edgecolor`

```python
edgecolor="0"
```

### What it does

* Sets the **color of the bar borders**

### Meaning of `"0"`

* `"0"` = **black**
* `"1"` = white
* Values between 0–1 give **shades of gray**

📌 `edgecolor="0"` → black borders

---

## 4️⃣ `errcolor`

```python
errcolor="0"
```

### What it does

* Sets the **color of the error bars**

📌 `"0"` again means **black error bars**

---

## Why your bars look like this

With:

```python
facecolor=(1,1,1,1)   # white bars
edgecolor="0"        # black border
linewidth=3          # thick border
errcolor="0"         # black error bars
```

👉 You get **white bars with thick black outlines and black error bars**


---

## Quick summary table

| Parameter | Controls         | Your Value |
| --------- | ---------------- | ---------- |
| linewidth | Border thickness | Thick (3)  |
| facecolor | Inside color     | White      |
| edgecolor | Border color     | Black      |
| errcolor  | Error bar color  | Black      |

---

## Exam tip 📝

📌 `(1,1,1,1)` = white (RGBA format)
📌 `"0"` = black, `"0.5"` = gray



````python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

boat = sns.load_dataset("titanic")
sns.set_palette("pastel")
sns.barplot(x="class", y="fare", data=boat,
            linewidth=3, facecolor=(1,1,1,1),
            errcolor="0.5", edgecolor="0")
plt.title("Boat Plot")
plt.show
````

````python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

boat = sns.load_dataset("titanic")
sns.set_palette("pastel")
sns.barplot(x="class", y="fare", data=boat,
            linewidth=3, facecolor=(0,0,0,1),
            errcolor="0", edgecolor="0")
plt.title("Boat Plot")
plt.show
````

## >>>Labelling
You can change the labels on x and y axis and also their
- colours
- font size
- font style

````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

boat = sns.load_dataset("titanic")
sns.boxplot(x="survived",y="age", data=boat, showmeans=True,
            meanprops= {"marker":"D",
                       "markersize":"15",
                       "markeredgecolor":"yellow"},)
plt.title("Titanic Plot", size=20, weight='bold')
plt.xlabel("How many survived", size=20, color="grey")
plt.ylabel("Age (Years)", size=20, color="red")
plt.show()
````

### Facet Grid and Wrap
Faceting mean splitting data into multiple subplots based on one or more categorical variables...
#### Facet Grid
Facet grid creates a grid of plots where,
- rows = one variable
- coloumns = another variable
- each cell contains the same type of plot (box plot)
- **Facet Grid = rows + columns**

````python
import seaborn as sns 
import matplotlib.pyplot as plt

boat = sns.load_dataset("titanic")
g = sns.FacetGrid(boat,
                 col="sex",
                 row="class")
g.map(sns.boxplot,"fare")
plt.show()
````

````python
import seaborn as sns 
import matplotlib.pyplot as plt

boat = sns.load_dataset("titanic")
g = sns.FacetGrid(boat,
                 col="class",
                 row="sex")
g.map(sns.boxplot,"fare")
plt.show()
````

````python
import seaborn as sns 
import matplotlib.pyplot as plt

boat = sns.load_dataset("titanic")
g = sns.FacetGrid(boat,
                 col="sex",
                 row="class")
g.map(sns.boxplot,"survived")
plt.show()
````

### Facet Wrap
Faceting means wrapping plots across coloumns when faceting by one variable
- col + col_wrap
- **Facet Wrap = Columns that wrap into rows**

````python
import seaborn as sns 
import matplotlib.pyplot as plt

boat = sns.load_dataset("titanic")
g = sns.FacetGrid(boat,
                 col="class",
                 col_wrap=3)
g.map(sns.boxplot,"fare")
plt.show()
````

````python
import seaborn as sns 
import matplotlib.pyplot as plt

boat = sns.load_dataset("titanic")
g = sns.FacetGrid(boat,
                 col="class",
                 col_wrap=3)
g.map(sns.boxplot,"sex")
plt.show()
````

````python
import seaborn as sns
sns.set_theme(style="whitegrid")

boat = sns.load_dataset("titanic")

# Draw a nested barplot by species and sex
g = sns.catplot(
    data=boat, kind="bar",
    x="class", y="survived", hue="sex",
    errorbar="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "embarked")
g.legend.set_title("")
````

We can plot different type of plots using the ready-made scripts.
- https://plotly.com/python/

````python
import plotly.express as px
# This dataframe has 244 lines, but 4 distinct values for `day`
df = px.data.tips()
fig = px.pie(df, values='tip', names='day')
fig.show()
````

````python
import plotly.express as px
df = px.data.gapminder().query("year == 2007").query("continent == 'Europe'")
df.loc[df['pop'] < 2.e6, 'country'] = 'Other countries' # Represent only large countries
fig = px.pie(df, values='pop', names='country', title='Population of European continent')
fig.show()
````

````python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Scatter(
    x=[1, 2, 3, 4], y=[10, 11, 12, 13],
    mode='markers',
    marker=dict(
        color=['rgb(93, 164, 214)', 'rgb(255, 144, 14)',
               'rgb(44, 160, 101)', 'rgb(255, 65, 54)'],
        opacity=[1, 0.8, 0.6, 0.4],
        size=[40, 60, 80, 100],
    )
)])

fig.show()
````

````python
import plotly.express as px
df = px.data.medals_long()

fig = px.area(df, x="medal", y="count", color="nation",
             pattern_shape="nation", pattern_shape_sequence=[".", "x", "+"])
fig.show()
````
