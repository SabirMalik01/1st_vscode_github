## Lists
- Ordered collection of elements
- Elements are enclosed in square braces [ ]
- Elements are Mutateable
  ### Functions on Lists
  | Method      | Usage                              |
| ----------- | ---------------------------------- |
| `append()`  | Add item at end                    |
| `insert()`  | Add item at specific index         |
| `pop()`     | Remove last item / item at index   |
| `remove()`  | Remove first occurrence of a value |
| `sort()`    | Sorts the list                     |
| `reverse()` | Reverses the list                  |
| `clear()`   | Empties the list                   |


```python
fruits = ["banana","orange","apple"]
fruits
```

```python
len(fruits)
```

```python
fruits[2]
```

```python
vegetables = ["Saag", "Palak", "Bhindi"]
vegetables
```

```python
favourites = fruits + vegetables
favourites
```

```python
fruits[2] = "grapes"
fruits
```

```python
fruits.count("a")
```

```python
fruits
```

```python
fruits.sort()
fruits
```

```python
list1 = [10,30,40,20,50]
list1
```

```python
list1.sort()
list1
```

```python
fruits
list1
vegetables
```

## Dictionaries
- An unordered collection of elements
- Elements are enclosed in curly braces { }
- Elements are stored as key:value
- Elements are mutable
  ### Functions on Dictionaries
  | Method     | Purpose                         |
| ---------- | ------------------------------- |
| `keys()`   | Returns all keys                |
| `values()` | Returns all values              |
| `items()`  | Returns all key–value pairs     |
| `get(key)` | Gets value safely without error |
| `update()` | Updates dictionary              |


```python
food1 = {"Burger":130,"Kabab":100, "Simosa":30}
food1
```

```python
food1.values()
```

```python
food1.keys()
```

```python
food1.items()
```

```python
food1.get("Burger")
```

```python
food1["Wings"]=400
food1
```

```python
food2 = {"Dates":300,"Kaju":600,"Akhrot":250}
food2
```

```python
food1.update(food2)
food1
```

```python
food2["Akhrot"] = 330
food2
```

```python
food2["Badaam"] = 500
food2
```

```python
food2["Kaju"]=600
food2
```

```python
food2
```

```python
food2.popitem()
food2
```

## Set
- Unordered and unindexed elements
- Curly braces { } are used
- Elements are not changeable
- Duplicate values are not allowed
  ### Functions on Set
  | Method           | Function                |
| ---------------- | ----------------------- |
| `add()`          | Add item                |
| `remove()`       | Remove item             |
| `discard()`      | Safe remove             |
| `clear()`        | Empty set               |
| `union()`        | Combine sets            |
| `intersection()` | Get common items        |
| `difference()`   | Items in set A not in B |



```python
set1 = {3.75, True, "Malik"}
set1
```

```python
set1.add("1st Sem")
set1
```

```python
set1
```

```python
set1.add("Programming")
set1
```

```python
set1.discard("Programming")
set1
```

```python
set2 = {3.53, "Bro", True}
set2
```

```python
set1.difference(set2)
set1
```

```python
set1.union(set2)
```

```python
set1.intersection(set2)
```

```python
set2.difference(set1)
```

```python
set1.difference(set2)
```

```python
set1.add("Work Hard Bro")
set1
```
